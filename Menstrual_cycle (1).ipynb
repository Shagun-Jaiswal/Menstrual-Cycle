### **Model Training**

```python
# Train test split
from sklearn.model_selection import train_test_split

X = data.drop(['ClientID', 'LengthofCycle', 'MeanCycleLength',"MeanMensesLength"],axis = 1)
y = data['LengthofCycle']

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.20, random_state = 20)
```

```python
print("Size of training data : ", X_train.shape)
print("Size of testing data : ", X_test.shape)
```

```python
X_train.shape[1]
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression().fit(X_train,y_train)
```

```python
y_pred_train=lr.predict(X_train)
y_pred_train
```

```python
y_pred_test = lr.predict(X_test)
y_pred_test
```

```python
plt.scatter(y_train, y_pred_train, s=30, alpha=0.7)

# Add a line representing perfect predictions
plt.plot(y_train, y_train, color='red')

# Set the plot title and axis labels
plt.title("Actual vs Predicted Values for Training Data")
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")

# Add a legend to the plot
#plt.legend(['Perfect Predictions', 'Model Predictions'])

# Show the plot
plt.show()
```

```python
plt.scatter(y_test, y_pred_test, s=30, alpha=0.7)

# Add a line representing perfect predictions
plt.plot(y_test, y_test, color='red')

# Set the plot title and axis labels
plt.title("Actual vs Predicted Values for Testing Data")
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")

# Add a legend to the plot
#plt.legend(['Perfect Predictions', 'Model Predictions'])

# Show the plot
plt.show()
```
